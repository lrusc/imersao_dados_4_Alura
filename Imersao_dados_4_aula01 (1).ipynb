```python
# desafio 1 :  Realizar a média da metragem para cara um dos bairros (Paulo)

metragem_media_bairros = dados.groupby(['Bairro']).agg({'Metragem':'mean'})
metragem_media_bairros
```

```python
# desafio 3:Explorar alguns gráficos na documentação e aplicar nas demais colunas do DF, tentar colocar alguma conclusão (Thiago)

# grafico de barras:
n_imoveis_bairro = dados.groupby(['Bairro']).agg({'Quartos':'mean'})
n_imoveis_bairro.head(10).plot.barh(title="Média de quarto por bairro", color="yellow")
```

```python
# grafico de área
n_imoveis_bairro = dados.groupby(['Bairro']).agg({'Vagas':'mean'})
n_imoveis_bairro.head(5).plot.area(title="Média de vagas por bairro", color="red")
```

```python
#gráfico de linha
n_imoveis_bairro = dados.groupby(['Bairro']).agg({'Banheiros':'mean'})
n_imoveis_bairro.head(5).plot.line(title="Média de banheiros por bairro", color="blue")
```

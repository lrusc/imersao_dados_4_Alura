```python
# desafio 1 :  Realizar a m√©dia da metragem para cara um dos bairros (Paulo)

metragem_media_bairros = dados.groupby(['Bairro']).agg({'Metragem':'mean'})
metragem_media_bairros
```

```python
# desafio 3:Explorar alguns gr√°ficos na documenta√ß√£o e aplicar nas demais colunas do DF, tentar colocar alguma conclus√£o (Thiago)

# grafico de barras:
n_imoveis_bairro = dados.groupby(['Bairro']).agg({'Quartos':'mean'})
n_imoveis_bairro.head(10).plot.barh(title="M√©dia de quarto por bairro", color="yellow")
```

```python
# grafico de √°rea
n_imoveis_bairro = dados.groupby(['Bairro']).agg({'Vagas':'mean'})
n_imoveis_bairro.head(5).plot.area(title="M√©dia de vagas por bairro", color="red")
```

```python
#gr√°fico de linha
n_imoveis_bairro = dados.groupby(['Bairro']).agg({'Banheiros':'mean'})
n_imoveis_bairro.head(5).plot.line(title="M√©dia de banheiros por bairro", color="blue")
```
